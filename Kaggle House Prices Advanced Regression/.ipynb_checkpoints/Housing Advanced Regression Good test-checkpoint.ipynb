```python
from sklearn.cross_validation import cross_val_score
from sklearn.metrics import make_scorer,mean_squared_error
```

```python
scorer = make_scorer(mean_squared_error,False)

def rmse_cv(model,X,y):
    return (cross_val_score(model,X,y,scoring=scorer)).mean()
```

```python
from sklearn.linear_model import Lasso
alphas = [1e-4,5e-4,1e-3,5e-3]
cv_lasso = (rmse_cv(Lasso(alpha=alpha,max_iter=50000),X_train,y_train)for alpha in alphas)
pd.Series(cv_lasso,index=alphas).plot()print("The MSE for Linear Regression(train) is :",mean_squared_error(y_train, pred))
print("The RMSE for Linear Regression(train) is :",np.sqrt(mean_squared_error(y_train,pred)))
```

```python
coef = pd.Series(model_lasso.coef_,index=X_train.columns).sort_values()
imp_coef = pd.concat([coef.head(10),coef.tail(10)])
imp_coef.plot(kind='barh')
plt.title('Coefficients in the Model')
```
