```python
import matplotlib.pyplot as plt
%matplotlib inline # lets us see the plots inside the Jupyter notebook, else use plt.show() if not using Jupyter Notebook
```

```python
import numpy as np
x = np.linspace(0,5,11)
```

```python
x
```

```python
y = x**2
```

```python
y
```

```python
# Functional Method : 
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object Oriented Method

fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes2.plot(x,y)
axes1.plot(y,x)
```
