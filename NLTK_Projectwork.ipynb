```python
# NLP Classification Task

# Create a dataframe called yelp_class that contains the columns of yelp dataframe 
# but for only the 1 or 5 star reviews.

yelp_class = yelp[(yelp.stars==1)|(yelp.stars==5)]
```

```python
yelp_class
```

```python
yelp_class.shape
```

```python
# Create two objects X and y. X will be the 'text' column of yelp_class 
# and y will be the 'stars' column of yelp_class. (Your features and target/labels)

X = yelp_class['text']
y = yelp_class['stars']
```

```python
# Import CountVectorizer and create a CountVectorizer object.

from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
X1 = cv.fit_transform(X)
```

```python
X1.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X1,y,test_size=0.3,random_state=101)
```

```python
X_test.shape
```

```python
#  We shall now train our model 
from sklearn.naive_bayes import MultinomialNB
```

```python
nb = MultinomialNB()
```

```python
nb.fit(X_train,y_train)
```

```python
prediction = nb.predict(X_test)
```

```python
#Evaluate the predictions via Classification report and confusion matrix
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,prediction))
print('\n')
print(confusion_matrix(y_test,prediction))
```

```python
# Let's see what happens if we try to include TF-IDF to this process using a pipeline.
# Using Text Processing
# Import TfidfTransformer from sklearn.

from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.pipeline import Pipeline
```

```python
pipeline = Pipeline([
    ('bow',CountVectorizer()), # Strings to token integer counts
    ('tfidf', TfidfTransformer()), # integer counts to weighted TF-IDF scores
    ('classifier',MultinomialNB()), # train on TF-IDF vectors w/ Naive Bayes classifier
])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
pipeline.fit(X_train,y_train)
```

```python
predictions = pipeline.predict(X_test)
```

```python
print(classification_report(y_test, predictions))
print('\n')
print(confusion_matrix(y_test,predictions))
```
