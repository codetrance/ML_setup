# Standard Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(bd.drop('Class',axis=1))
```

```python
scaled_features = scaler.fit_transform(bd.drop('Class',axis=1))
```

```python
bd_feat = pd.DataFrame(scaled_features,columns=bd.columns[:-1])
```

```python
bd_feat.head()
```

```python
X = bd_feat
y = bd['Class']
```

```python
X = X.as_matrix()
y = y.as_matrix()
```
